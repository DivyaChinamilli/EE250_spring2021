```python
%%writefile lab2_test.py
#!/usr/bin/python3

import os
import sys
import random
import base64
import unittest
import subprocess

TIMEOUT = 20

REF_SOLUTION = b''


class Lab2TestClient(unittest.TestCase):

  def setUp(self):
    """
    Compile the program before testing
    """
    subprocess.check_call(["make", "lab2"])

    with open("lab2-ref.o", "wb") as f:
      f.write(base64.b64decode(REF_SOLUTION))
      
    subprocess.check_call(["g++", "lab2-ref.o", "-o", "lab2-ref.out"])
       

  def run_programs(self, delay):

    p = subprocess.Popen(["./lab2-ref.out", "server", f"{delay}"], stdout=subprocess.PIPE, text=True)
    r = subprocess.run(["./lab2.out", "client"], timeout=TIMEOUT)
    try:
      p.wait(timeout=TIMEOUT)
    except:
      p.terminate()
    
    return [l.strip() for l in p.stdout.readlines()]

  
  def test_connection(self):

    delay = random.randint(10, 19)

    user_result = self.run_programs(delay)

    self.assertIn("ACCEPTED", user_result)

  
  def test_discovery(self):

    delay = random.randint(10, 19)

    user_result = self.run_programs(delay)

    self.assertIn("DISCOVERY: OK", user_result)

  
  def test_request(self):

    delay = random.randint(10, 19)

    user_result = self.run_programs(delay)

    self.assertIn("REQUEST: OK", user_result)


  def test_resultfile(self):
    delay = random.randint(10, 19)

    self.run_programs(delay)

    with open("result-client.txt") as f:
      datalines = f.readlines()

    user_delay, user_ip = datalines[0].strip().split()

    user_delay = int(user_delay)

    self.assertLess(abs(delay * 1000 - user_delay), 2000)
    self.assertEqual(f"10.0.0.{delay}", user_ip)


class Lab2TestServer(unittest.TestCase):

  def setUp(self):
    """
    Compile the program before testing
    """
    subprocess.check_call(["make", "lab2"])

    with open("lab2-ref.o", "wb") as f:
      f.write(base64.b64decode(REF_SOLUTION))
      
    subprocess.check_call(["g++", "lab2-ref.o", "-o", "lab2-ref.out"])
       

  def run_programs(self, tid):

    p = subprocess.Popen(["./lab2.out", "server"])
    r = subprocess.run(["./lab2-ref.out", "client", f"{tid}"], timeout=TIMEOUT, stdout=subprocess.PIPE, text=True)
    try:
      p.wait(timeout=TIMEOUT)
    except:
      p.terminate()
    
    return [l.strip() for l in r.stdout.split("\n")]

  
  def test_connection(self):

    tid = random.randint(100, 199)

    user_result = self.run_programs(tid)

    self.assertIn("CONNECTED", user_result)

  
  def test_discovery(self):

    tid = random.randint(100, 199)

    user_result = self.run_programs(tid)

    self.assertIn("OFFER: OK", user_result)

  
  def test_request(self):

    tid = random.randint(100, 199)

    user_result = self.run_programs(tid)

    self.assertIn("ACKNOWLEDGE: OK", user_result)


  def test_resultfile(self):
    tid = random.randint(100, 199)

    self.run_programs(tid)

    with open("result-server.txt") as f:
      datalines = f.readlines()

    user_tid, user_ip = datalines[0].strip().split()

    user_tid = int(user_tid)

    self.assertEqual(tid, user_tid)

  
if __name__ == '__main__':

  suite = unittest.TestSuite()

  if len(sys.argv) == 2 and sys.argv[1] == "test-client":
    
    suite.addTest(unittest.defaultTestLoader.loadTestsFromTestCase(Lab2TestClient))


  elif len(sys.argv) == 2 and sys.argv[1] == "test-server":
    
    suite.addTest(unittest.defaultTestLoader.loadTestsFromTestCase(Lab2TestServer))


  elif len(sys.argv) == 2 and sys.argv[1] == "test-all":

    suite.addTest(unittest.defaultTestLoader.loadTestsFromTestCase(Lab2TestClient))
    suite.addTest(unittest.defaultTestLoader.loadTestsFromTestCase(Lab2TestServer))


  else:
    print("No tests requested")
    sys.exit(1)


  result = unittest.TextTestRunner().run(suite)

  # Set the exit code based on the test result
  sys.exit(not result.wasSuccessful())
```
